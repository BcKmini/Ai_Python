```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
x1 = np.array([2, 3, 5, 10])
y2 = np.array([5, 3, 7, 8])

plt.plot(x, x1, "ro-")  # Using 'x1' instead of 'y1'
plt.plot(x, y2, "go-.")
plt.show()
```

```python
#sin, cos 곡선 그리기

x = np.arange(0, 3*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2, "g:")
plt.show()
```

```python
#산점도 -> scatter
p = np.random.rand(100,2)

plt.scatter(p[:,0], p[:,1])
plt.show()
```

```python
#pie 그래프 그려보기 
labels = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = [0.3, 0, 0, 0, 0, 0]

plt.pie(sizes,
        explode = explode,
        labels=labels,
        autopct='%1.1f%%', #비율, 소수점 1자리까지,
        shadow=True, # 파이조각에 그림자,
        startangle=90, # 시작에 각도가 90도부터 시작
        )
plt.show()
```

```python
#그래프 연습해보기 
plt.rc('font', family="Malgun Gothic")

plt.title("연습 그래프", fontsize=20)
plt.plot([10,20,30,40], [1,4,9,16],
         c="b", lw=5, ls="--",
         marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.show()
```

```python
t = np.arange(0, 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
#ex1
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 5.0)

y1 = np.cos(2 * np.pi * x1 ) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2) 


fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Title')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel("Damped oscilltion")

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('Undamped')

plt.show()
```

```python
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(5,5))
axs[0, 0 ].hist(data[0])
axs[1, 0 ].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```

```python
x_label = ['a', 'b', 'c', 'd', 'e', 'f']
x = np.arange(len(x_label))
data = np.random.random_sample(size=(6,3))
y1 = data[:, 0]
y2 = data[:, 1]
#plot
fig, ax = plt.subplots()

#set width
width = 0.35
ax.bar(x - width/2, y1, width, align='center')
ax.bar(x + width/2, y2, width, align='center')
ax.legend(['ax1', 'ax2'])
plt.show()
```
