Indexing & Slicing -> DataFrame 조작해보기

```python
import pandas as pd

# 데이터 생성
data = {
    'Name': ['Lee', 'Kim', 'Choi', 'Kim'],
    'year': [2013, 2014, 2015, 2016],
    'points': [1.5, 1.6, 1.7, 1.8]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 'year' 열 출력
print(df['year'], '\n')

# 'Name'과 'points' 열 출력
print(df[['Name', 'points']], '\n')

# 인덱스 1부터 2까지의 행 출력 (1:3이므로 1, 2행 출력)
print(df.iloc[1:3], '\n')
```

```python
import pandas as pd
import numpy as np

# DatetimeIndex 생성
dates = pd.date_range('20240715', periods=6)

# DataFrame 생성
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
print(df, '\n')

# Series 생성 및 DataFrame에 추가
sl = pd.Series([1, 2, 3, 4, 5, 6], index=dates)
df['F'] = sl

# 특정 위치의 값 수정
df.at[dates[0], 'A'] = 0

# 열 D의 값을 모두 5로 설정
df.loc[:, 'D'] = np.array([5] * len(df))

# DataFrame의 음수 값을 양수로 변환
df[df < 0] = -df

print(df)
```
