```python
import numpy as np
from scipy.linalg import *

a = np.array([[1,2,3], [5,7,11], [21,29,31]])

print(a, '\n')
print(f"det(a) = {det(a)}")
print(f"inv(a) =\n {inv(a)}")
print(f"a * inv(a) =\n{np.dot(a, inv(a))}")
```

```python
a = np.array([[4, 2], [3, 5]])
print(f"a =\n{a}")

eigenvalues, eigenvectors = eig(a)
print(f"eigenvalues =\n{eigenvalues}")
print(f"eigenvectors =\n{eigenvectors}")
```

```python
'''
연립방정식 풀기
3x + 2y = 2
x - y = 4
5y + z = -1
'''
A = np.array([[3,2,0], [1,-1,0], [0,5,1]])
B = np.array([2,4,-1]) #열벡터
s = solve(A, B)
print(f"(x,y,z) = {s}")
print(f"A * s = {np.dot(A, s)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mean, sigma):
    return (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(- (x - mean)**2 / (2 * sigma**2))

labels = []
x = np.arange(-5, 5, 0.01)
for i in range(1, 5):
    labels.append(f"sigma = {i}")
    plt.plot(x, gaussian(x, 0, i))
plt.grid()
plt.legend(labels)
plt.show()
```

```python
#최적화란 -> 미분 가능한 함수의 최저점을 찾는 문제
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

def f1(x):
    return ( x - 2) ** 2 + 2

xx = np.linspace(-1, 4, 100)
plt.plot(xx, f1(xx))
plt.plot(2, 2, 'ro', markersize=10)
plt.ylim(0, 10)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("1차원 목적함수")
plt.show()
```

```python
#미분값이 '0'이되는 순간
import numpy as np
from scipy.optimize import fmin_bfgs

def f(x):
    return x**2 + 10*np.sin(x)

print(fmin_bfgs(f, 0))
print(fmin_bfgs(f, 6))
```

```python
import plotly.express as px

# Iris 데이터셋 로드
df = px.data.iris()

# 3D 산점도 생성
fig = px.scatter_3d(df, x='sepal_width', y='sepal_length', z='petal_width', color='species')
fig.update_traces(marker_size=1.2)

# HTML 파일로 저장
fig.write_html("iris_3d_scatter.html")

# 저장된 HTML 파일을 브라우저에서 열기
import webbrowser
webbrowser.open("iris_3d_scatter.html")
```
