#relaton plot 실습

```python
tips = sns.load_dataset('tips')
sns.set_theme(style="darkgrid", context='notebook')
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
#계열화_그룹핑

sns.set_theme(style='darkgrid')
sns.relplot(x='total_bill', y='tip', hue='smoker', style='time', size='size', data=tips)
plt.show()
```

```python
iris
```

```python
sns.set_theme(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', hue='species', kind='line', data=iris, errorbar=None)
plt.show()
#연한색은 평균과 sns간의 95%신뢰구간 자동으로 표시됨
```

```python
iris = sns.load_dataset('iris')

sns.set_theme(style="darkgrid")
sns.displot(data=iris, x="petal_length", binwidth=0.3, hue='species', multiple="dodge")
sns.displot(data=iris, x='petal_length', hue='species', multiple="stack")
sns.displot(data=iris, x='petal_length', hue='species', multiple="fill")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.displot(iris, x='petal_length', hue="species", kde = True)
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.displot(iris, x='petal_length', hue="species", kind="kde")
plt.show()
```

이변량분포

```python
sns.set_theme(style="darkgrid")
sns.displot(iris, x="petal_length", y="sepal_length", hue="species", kind="kde")
plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.set_theme(style="darkgrid")
sns.catplot(data=titanic, x='class', hue='who', kind='count')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.catplot(data=tips, x='size', hue='day', kind='count')
plt.show()
```

```python
sns.set(style="darkgrid")
sns.lmplot(x= 'total_bill', y='tip', col='time', hue="smoker", data=tips)
plt.show()
```

pairplot

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

jointgplot

```python
sns.jointplot(data=iris, x='petal_length', y='sepal_length', hue='species')
plt.show()
```

FacetGrid로 다중 Plot 그리기 

```python
g = sns.FacetGrid(tips, col='day')
g.map(sns.histplot, 'total_bill')
plt.show()
```
