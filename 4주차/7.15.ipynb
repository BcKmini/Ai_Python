EDA(Exploratory Data Analysis)란? -> 탐색적 데이터 분석

❖  데이터 분석의 종류

▪  탐색적 데이터 분석(EDA: Exploratory Data Analysis)

• 가장 기초적인 데이터분석으로 그래프 분석이라고도 함
    − 일변량 분석(확률분도포, 시계열분석    등)
    − 이변량 분석(산점도, 회귀분석, 논리적회귀분석    등)
    − 다변량 분석(Matrix plot, Contour plot 등)
▪ 정성적 데이터 분석(QDA: Qualitative Data Analysis)
• 경험과 노하우, 브레인스토밍을 통해 의사 결정을 하는 분석법

▪ 빅데이터 분석(정량적 데이터 분석, R분석) -> R분석 사용해보기
• 시장 데이터나 사내 서버에 저장된 대용량 데이터 분석

▪ 확증적 데이터 분석(CDA: Confirmatory Data Analysis)
• 설정한 가설과 일치하는지 그렇지 않은지를 판단하는 '가설 검증'이 주류를 이루고 있음

EDA의 필요성

❖  EDA는 데이터를 시각화 및 변환하여 직관적이고 체계적인 방법으로 탐구하는 방법 및 활동을 의미

❖  EDA의 필요성
▪ 데이터의 분포 및 값을 여러 각도에서 관찰함으로써, 데이터의 전체적인 양상과 보이지 않던 현상을 잘 이해할 수 있음
• 데이터 준비단계에서 미처 발견하지 못한 잠재적 문제들 발견할 수 있도록 함
• 문제 정의단계에서  미처 발견하지 못한 다양한 패턴을 발견함으로써 기존의 가설을 수정하거나 새로운 가설을 추가할 수 있음

EDA의 과정(중요**)

1. 분석의 목적 확인
2. 어떠한 변수가 존재하는지 확인(변수의 이름, 설명 등)
3. 데이터를 전체적으로 살펴보기
▪  데이터에 문제가 없는지 확인
▪  head 또는 tail부분을 확인
4. 속성간의 관계 분석
▪ Categorical과 Numerical 속성의 조합
5. 이상치나 NULL(NaN)과 같은 데이터의 존재유무 확인
▪ 적절한 통계 요약지표(평균, 중간값, 분산 등)
▪ 시각화 도구(확률밀도, 히스토그램, 시계열챠트 등)
▪ 기계학습 기법(k-means 알고리즘 등)

4. jupyter notebook markdown

(1) Basic of Markdown

    1. Header
    ▪ H1 ~ H6까지    태그를    #으로    표현

    # 헤더 스타일 (h1)
    ## 헤더 스타일 (h2) 
    ### 헤더 스타일 (h3) 
    #### 헤더 스타일 (h4) 
    ##### 헤더 스타일 (h5) 
    ###### 헤더 스타일 (h6)

    2. 수평선
    '-', '*', '_' 문자를    세    개    이상    나열하면    수평선
    단락을    나눌    때    사용

    3. 텍스트 강조
    *single asterisks* - 이탤릭체
    _single underscores_ - 이탤릭체
    **double asterisks** - 볼드체
    **_double underscores_** - 볼드이탤릭체 
    ~~cancelline~~ - 취소선


▪  MarkDown으로 숫자 리스트, 블릿 리스트를 표현할 수 있음.  
▪  순차가 없는 리스트를 표현할 때는 ‘*‘, ‘+’, ‘-” 기호 사용 가능.  
▪  순차가 있는 리스트를 표현할 때 ‘숫자’ 사용 가능  
▪  레벨을 들여 쓸때에는 tab으로...  
▪  만일, 특정 레벨에서 문단을 나누는 경우에는 문장이 시작하는 곳에 공백문자를 1~2개 정도 넣어준다.

❖  코드블럭    & 인라인    코드
▪  코드    첫라인    앞과    마지막    라인    다음에    ` (fense)문자로    둘러싼다.
▪  단    fense문자가    3개씩이면    코드블럭, 1개씩이면    인라인코드가    된다.
▪  코드블럭에    해당    코드언어명(예: java, json, python)을    적어주면    됨.


- 코드블럭
``` python
x = 10
y = x ** 2 print(y) 
```
- 인라인    코드는    `n = input('input a number : ')`와    같은    것을    말한다.
