### 2) 모델 구조만 빌려와서 사용해보기 (권장하지 않음)

* 모델 구조 선택하기

```python
model_scratch = YOLO(model='yolov8n.yaml', task='detect')
```

* 모델 학습하기

```python
model_scratch.train(data='/content/futbol-players-7/data.yaml',
                    epochs=1,
                    patience=5,
                    seed=2024,
                    pretrained=False,
                    )
```

* 예측해보기

### 3) 모델 구조에 사전 학습 가중치도 가져와서 사용해보기

* 모델 구조 및 사전 학습 가중치 선택하기

```python
model_transfer = YOLO(model='yolov8n.pt', task='detect')
```

* 모델 학습하기

```python
model_transfer.train(model='/content/yolov8n.pt',
                    data='/content/futbol-players-7/data.yaml',
                    epochs=20,
                    patience=5,
                    seed=2024,
                    pretrained=True,
                    )
```

* 예측해보기
