##**2. Streamlit ì½”ë”©**

```python
# ì´ íŒŒì¼ì— Streamlit ì• í”Œë¦¬ì¼€ì´ì…˜ ì½”ë“œë¥¼ ì‘ì„±
# ì½”ë“œ ìˆ˜ì • í›„, ì…€ì„ ì‹¤í–‰í•´ì•¼ ë³€ê²½ëœ ë‚´ìš©ì´ ooo.py íŒŒì¼ì´ ì ìš©ë¨
%%writefile /content/drive/MyDrive/Streamlit/4-1.input.py
# streamlit, pandas ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import streamlit as st
import pandas as pd

st.title('Unit 4. Input Widgets')
st.caption('ì°¸ì¡°ì‚¬ì´íŠ¸: https://docs.streamlit.io/library/api-reference/widgets')

st.header('1. Button')
if st.button('Say hello'):
    st.write('Hello')
else:
    st.write('Goodbye')

st.header('2. Radio button')
genre = st.radio('ì¢‹ì•„í•˜ëŠ” ì˜í™” ì¥ë¥´ë¥¼ ì„ íƒí•˜ì„¸ìš”',('ì½”ë©”ë””', 'SF', 'ì•¡ì…˜'))

if genre == 'ì½”ë©”ë””':
    st.write('ì½”ë©”ë”” ìœ ì¾Œí•˜ì‹  ë¶„ì´ì‹œêµ°ìš”')
elif genre == 'SF':
    st.write('ì €ë„ SF ì¢‹ì•„í•©ë‹ˆë‹¤')
else:
    st.write('ë©‹ì§€ì‹­ë‹ˆë‹¤.')

st.header('3. Checkbox')
agree = st.checkbox('I agree')
if agree:
    st.write('ğŸ˜„'*10)

st.header('4. toggle')
on = st.toggle('Activate toggle')
if on:
    st.write('Feature activated!')

st.header('5. Select box')
option = st.selectbox('ì–´ë–»ê²Œ ì—°ë½ ë“œë¦´ê¹Œìš”?',('Email', 'Mobile phone', 'Office phone'))
st.write('ë„¤ ', option, ' ì˜ ì•Œê² ìŠµë‹ˆë‹¤')

st.header('6. Multi select')
options = st.multiselect('ì¢‹ì•„í•˜ëŠ” ìƒ‰ì„ ëª¨ë‘ ì„ íƒí•˜ì„¸ìš”',
         ['Green', 'Yellow', 'Red', 'Blue'],
         ['Yellow', 'Red'])

st.write('ì„ í˜¸ ìƒ‰ìƒ:', ', '.join(options))
```
